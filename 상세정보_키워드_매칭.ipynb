```python
from transformers import pipeline

pipe = pipeline("zero-shot-classification", model="pongjin/roberta_with_kornli")
```

```python
import pandas as pd

datas = pd.read_csv('14_상세정보_1.csv')

sequences = datas['overview'].head()
print(sequences)
```

```python
pipe = pipeline("zero-shot-classification", model="pongjin/roberta_with_kornli")
candidate_labels =['가족','문화', '연인', '친구', '역사','자연','교육']

sequence = datas['overview'][13]
#print(sequence)
re = pipe(sequence,candidate_labels,hypothesis_template='이는 {}에 관한 것이다.',)
re
```

```python
df_all = pd.DataFrame()
candidate_labels =['가족','문화', '연인', '친구', '역사','자연']
for sequence in sequences:
  try:

    pipe = pipeline("zero-shot-classification", model="pongjin/roberta_with_kornli")
    re = pipe(
    sequence,
    candidate_labels,
    hypothesis_template='이는 {}에 관한 것이다.',
    )
    print(re)
    df = pd.json_normalize(re)
    df_all = pd.concat([df,df_all],ignore_index=True)
  except:
    continue


df_all.to_csv('result.csv', encoding='utf-8-sig')
df_all
```
