```python
# 데이터 불러오기 test


import requests
import pandas as pd
import json

url = 'http://apis.data.go.kr/B551011/KorWithService1/areaBasedSyncList1?serviceKey=servicekey&numOfRows=500&pageNo=4&MobileOS=ETC&MobileApp=AppTest&_type=json&showflag=1&listYN=Y&arrange=C&contentTypeId=12'
res = requests.get(url)
contents = res.text

json_ob = json.loads(contents)
#print(json_ob)

body = json_ob['response']['body']['items']['item'] #json 특정요소만 불러오기
df = pd.json_normalize(body)

result = df.to_csv('result.csv', encoding='utf-8-sig')

# inplace=True 해두면 영구적으로 수정되는 것임

csv = pd.read_csv('result.csv')
#csv = csv.drop(labels=['addr2','booktour','cat1','cat2','cat3','cpyrhtDivCd','mapx','tel'], axis=1)
contentIds = csv['contentid'].values
contentIds = contentIds.tolist()
contentIds.pop(1)
print(contentIds)
csv
```

```python
# 코드 별로 상세정보 받아오기

df_all=pd.DataFrame()

for contentId in contentIds:
  url = f"http://apis.data.go.kr/B551011/KorWithService1/detailCommon1?serviceKey=servicekey&MobileOS=ETC&MobileApp=AppTest&_type=Json&contentId={contentId}&defaultYN=Y&overviewYN=Y"
  res = requests.get(url)
  contents = res.text
  try:
    json_ob = json.loads(contents)


    #print(json_ob)
      # json 특정요소만 불러오기
    body = json_ob['response']['body']['items']['item']


    df = pd.json_normalize(body)



      # 데이터를 리스트에 추가
    df_all = pd.concat([df, df_all], ignore_index=True)
  except:
    continue


df_all.to_csv('result.csv', encoding='utf-8-sig')
#csv = csv.drop(labels=['contenttypeid','booktour','createdtime','modifiedtime','tel','booktour'], axis=1, inplace=True)

df_all['title']
```
