```python
from konlpy.tag import Komoran
import pandas as pd
from textrank import KeywordSummarizer

# CSV 파일 로드
csv_file_path = '12_상세정보_1.csv'  # 파일 경로를 입력하세요
df = pd.read_csv(csv_file_path)

# Komoran 형태소 분석기 초기화
komoran = Komoran()

# 한글 불용어 리스트
korean_stop_words = [
    '그리고', '그', '저', '것', '이', '그것', '저것', '그러한', '저러한', '등', '때문','가장','다시',
    '이것', '이다', '있다', '없다', '한다', '있습니다', '있어서', '입니다', '하는', '주변','고기','아래', '등록','고여','마련',
    '하여', '한', '하지', '보다', '하고', '해서', '했다', '했습니다', '할', '사람', '의자', '방문','달이','수','마을','만끽','월','정리'
]

# Komoran 토크나이저 함수 정의
def komoran_tokenizer(sent):
    words = komoran.pos(sent, join=True)
    # 명사(NN)만 필터링 후 불용어 제거
    words = [w for w in words if '/NN' in w and w.split('/')[0] not in korean_stop_words]
    return words

# 첫 번째 행부터 50번째 행까지 반복 수행
for index, row in df.head(50).iterrows():
    # 각 행의 'overview' 열에서 텍스트 추출 및 Komoran으로 토큰화
    sents = row['overview']
    tokenized_words = komoran_tokenizer(sents)

    # 토큰화된 단어를 하나의 문자열로 결합
    converted_string = [' '.join(tokenized_words)]

    # KeywordSummarizer 초기화 및 키워드 요약 수행
    keyword_summarizer = KeywordSummarizer(tokenize=komoran_tokenizer, min_count=2, min_cooccurrence=1)
    keywords = keyword_summarizer.summarize(converted_string, topk=15)

    # 각 행의 Content ID와 키워드 출력
    print(f"Content ID: {row['contentid']}")
    print(f"Keywords:")
    for word, rank in keywords:
        print(f'  {word} ({rank:.3})')

    print("-" * 40)

    # 키워드에서 형태소 정보 제거하고 단어만 추출
    words = [item[0].split('/')[0] for item in keywords]
    print(f"Filtered Keywords: {words}")
```

## word2vec + consine similarity

```python
from gensim.models import Word2Vec
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# 주어진 input 단어 리스트
input_words = [
    '월', '정리', '해변', '바다', '사람', '제주도', '여행객', '수', '마을', '달', '풍경', '방문', '서핑',
    '만끽', '때문'
]

# 키워드 리스트
keywords = [
    '투어', '숲', '힐링', '모험', '박물관', '미술관', '하이킹', '트레킹', '캠핑', '피크닉', '시장',
    '쇼핑', '기념품', '공원', '놀이공원', '테마파크', '이벤트', '축제', '공연', '랜드마크', '관광',
    '레저', '애견', '낚시', '스포츠'
]

# Word2Vec 모델 로드 (사전 훈련된 모델을 사용하거나 직접 훈련)
# 여기서는 직접 훈련을 가정
sentences = [input_words + keywords]
model = Word2Vec(sentences, vector_size=100, window=5, min_count=1, workers=4)

# 각 단어를 벡터로 변환
input_vectors = np.array([model.wv[word] for word in input_words])
keyword_vectors = np.array([model.wv[word] for word in keywords])

# 각 input 단어에 대해 키워드와의 코사인 유사도 계산 및 상위 3개 키워드 출력
results = {}
for i, input_word in enumerate(input_words):
    cosine_similarities = cosine_similarity(input_vectors[i:i+1], keyword_vectors).flatten()
    top_indices = cosine_similarities.argsort()[-3:][::-1]
    top_keywords = [keywords[idx] for idx in top_indices]
    results[input_word] = top_keywords

results
```

## FastText + consine similarity

```python
from gensim.models import FastText
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# 주어진 input 단어 리스트
input_words = [
    '월', '정리', '해변', '바다', '사람', '제주도', '여행객', '수', '마을', '달', '풍경', '방문', '서핑',
    '만끽', '때문'
]

# 키워드 리스트
keywords = [
    '투어', '숲', '힐링', '모험', '박물관', '미술관', '하이킹', '트레킹', '캠핑', '피크닉', '시장',
    '쇼핑', '기념품', '공원', '놀이공원', '테마파크', '이벤트', '축제', '공연', '랜드마크', '관광',
    '레저', '애견', '낚시', '스포츠'
]

# FastText 모델 훈련
sentences = [input_words + keywords]
model = FastText(sentences, vector_size=100, window=5, min_count=1, workers=4)

# 각 단어를 벡터로 변환
input_vectors = np.array([model.wv[word] for word in input_words])
keyword_vectors = np.array([model.wv[word] for word in keywords])

# 각 input 단어에 대해 키워드와의 코사인 유사도 계산 및 상위 3개 키워드 출력
results = {}
for i, input_word in enumerate(input_words):
    cosine_similarities = cosine_similarity(input_vectors[i:i+1], keyword_vectors).flatten()
    top_indices = cosine_similarities.argsort()[-3:][::-1]
    top_keywords = [keywords[idx] for idx in top_indices]
    results[input_word] = top_keywords

results
```

## BERT + cosine similarity

```python
import torch
from transformers import BertTokenizer, BertModel
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# BERT 모델 및 토크나이저 로드
tokenizer = BertTokenizer.from_pretrained('bert-base-multilingual-cased')
model = BertModel.from_pretrained('bert-base-multilingual-cased')

# 주어진 input 단어 리스트
input_words = [
    '월', '정리', '해변', '바다', '사람', '제주도', '여행객', '수', '마을', '달', '풍경', '방문', '서핑',
    '만끽', '때문'
]

# 키워드 리스트
keywords = [
    '투어', '숲', '힐링', '모험', '박물관', '미술관', '하이킹', '트레킹', '캠핑', '피크닉', '시장',
    '쇼핑', '기념품', '공원', '놀이공원', '테마파크', '이벤트', '축제', '공연', '랜드마크', '관광',
    '레저', '애견', '낚시', '스포츠'
]

# 단어들을 BERT 임베딩 벡터로 변환하는 함수 정의
def get_bert_embeddings(words):
    inputs = tokenizer(words, return_tensors='pt', padding=True, truncation=True, max_length=10)
    with torch.no_grad():
        outputs = model(**inputs)
    return outputs.last_hidden_state.mean(dim=1).numpy()

# input 단어들과 키워드들을 BERT 임베딩 벡터로 변환
input_embeddings = get_bert_embeddings(input_words)
keyword_embeddings = get_bert_embeddings(keywords)

# 각 input 단어에 대해 키워드와의 코사인 유사도 계산 및 상위 3개 키워드 출력
results = {}
for i, input_word in enumerate(input_words):
    cosine_similarities = cosine_similarity(input_embeddings[i:i+1], keyword_embeddings).flatten()
    top_indices = cosine_similarities.argsort()[-3:][::-1]
    top_keywords = [keywords[idx] for idx in top_indices]
    results[input_word] = top_keywords

# 결과 출력
for word, matched_keywords in results.items():
    print(f"단어: {word}, 매칭된 키워드: {matched_keywords}")
```
