```python
import pandas as pd
from konlpy.tag import Okt
from sklearn.feature_extraction.text import TfidfVectorizer

# Okt 형태소 분석기 초기화
okt = Okt()

# CSV 파일 로드
csv_file_path = '12_상세정보_1.csv'  # 파일 경로를 입력하세요
df = pd.read_csv(csv_file_path)

df = df[['contentid','overview']].head() # 필요한 컬럼만 추출


# 명사 추출 함수
def extract_nouns(text):
    nouns = okt.nouns(text)
    return ' '.join(nouns)

# 'overview' 열에서 명사만 추출
df['nouns'] = df['overview'].apply(extract_nouns)

# TF-IDF 벡터라이저 초기화
vectorizer = TfidfVectorizer()

# TF-IDF 행렬 생성
tfidf_matrix = vectorizer.fit_transform(df['nouns'])


# 단어 리스트 생성
feature_names = vectorizer.get_feature_names_out()
#print(feature_names)

# 유의미한 명사 3개 추출 함수
def get_top_n_words(tfidf_row, feature_names, top_n=3):
    row = tfidf_row.toarray().flatten()
    top_indices = row.argsort()[-top_n:][::-1]
    top_words = [feature_names[i] for i in top_indices]
    return top_words

# 각 문서에서 상위 3개의 유의미한 명사 추출
df['top_nouns'] = [get_top_n_words(tfidf_matrix[i], feature_names) for i in range(tfidf_matrix.shape[0])]

# 결과 출력
df[['overview', 'top_nouns']]
```

```python
import pandas as pd
from konlpy.tag import Okt
import networkx as nx

# Okt 형태소 분석기 초기화
okt = Okt()

# CSV 파일 로드
csv_file_path = '12_상세정보_1.csv'  # 파일 경로를 입력하세요
df = pd.read_csv(csv_file_path)
df = df[['contentid','overview']].head() # 필요한 컬럼만 추출

# 명사 추출 함수
def extract_nouns(text):
    return okt.nouns(text)

# TextRank 알고리즘을 사용한 키워드 추출 함수
def textrank_keyword_extraction(text, top_n=3):
    # 명사 추출
    words = extract_nouns(text)

    # 그래프 생성
    graph = nx.Graph()

    # 명사 간의 연결 생성
    for i, word in enumerate(words):
        for j in range(i + 1, len(words)):
            if j - i > 10:  # 너무 먼 단어끼리는 연결하지 않음
                break
            graph.add_edge(word, words[j])

    # PageRank 계산
    ranks = nx.pagerank(graph)

    # 상위 키워드 추출
    keywords = sorted(ranks, key=ranks.get, reverse=True)[:top_n]
    return keywords

# 각 문서에서 주요 키워드 추출
df['keywords'] = df['overview'].apply(lambda x: textrank_keyword_extraction(x, top_n=3))

# 결과 출력
print(df[['overview', 'keywords']])
```

```python
from collections import Counter
def scan_vocabulary(sents, tokenize, min_count=2):
    # 각 문장을 토큰화하여 단어 빈도수 카운팅
    counter = Counter(w for sent in sents for w in tokenize(sent))
    # min_count 이상 나타난 단어만 선택
    counter = {w: c for w, c in counter.items() if c >= min_count}
    # 빈도수 순으로 정렬
    idx_to_vocab = [w for w, _ in sorted(counter.items(), key=lambda x: -x[1])]
    # 단어를 인덱스에 매핑
    vocab_to_idx = {vocab: idx for idx, vocab in enumerate(idx_to_vocab)}
    return idx_to_vocab, vocab_to_idx



from collections import defaultdict
from scipy.sparse import csr_matrix

def dict_to_mat(d, n_rows, n_cols):
    rows, cols, data = [], [], []
    for (i, j), v in d.items():
        rows.append(i)
        cols.append(j)
        data.append(v)
    return csr_matrix((data, (rows, cols)), shape=(n_rows, n_cols))

def cooccurrence(tokens, vocab_to_idx, window=2, min_cooccurrence=2):
    counter = defaultdict(int)
    for s, tokens_i in enumerate(tokens):
        vocabs = [vocab_to_idx[w] for w in tokens_i if w in vocab_to_idx]
        n = len(vocabs)
        for i, v in enumerate(vocabs):
            if window <= 0:
                b, e = 0, n
            else:
                b = max(0, i - window)
                e = min(i + window + 1, n)
            for j in range(b, e):
                if i == j:
                    continue
                counter[(v, vocabs[j])] += 1
                counter[(vocabs[j], v)] += 1
    counter = {k: v for k, v in counter.items() if v >= min_cooccurrence}
    n_vocabs = len(vocab_to_idx)
    return dict_to_mat(counter, n_vocabs, n_vocabs)


def word_graph(sents, tokenize=None, min_count=2, window=2, min_cooccurrence=2):
    idx_to_vocab, vocab_to_idx = scan_vocabulary(sents, tokenize, min_count)
    tokens = [tokenize(sent) for sent in sents]
    g = cooccurrence(tokens, vocab_to_idx, window, min_cooccurrence)
    return g, idx_to_vocab



```

```python
import numpy as np
from sklearn.preprocessing import Normalizer

def pagerank(x, df=0.85, max_iter=30):
    assert 0 < df < 1

    # initialize
    A = normalize(x, axis=0, norm='l1')
    R = np.ones(A.shape[0]).reshape(-1, 1)
    bias = (1 - df) * np.ones(A.shape[0]).reshape(-1, 1)

    # iteration
    for _ in range(max_iter):
      R = df * (A * R) + bias

    return R

```

```python
def textrank_keyword(sents, tokenize, min_count, window, min_cooccurrence, df=0.85, max_iter=30, topk=10):
    g, idx_to_vocab = word_graph(sents, tokenize, min_count, window, min_cooccurrence)
    R = pagerank(g, df, max_iter).reshape(-1)
    idxs = R.argsort()[-topk:]
    keywords = [(idx_to_vocab[idx], R[idx]) for idx in reversed(idxs)]
    return keywords
```

```python
from konlpy.tag import Komoran
import pandas as pd

csv_file_path = '12_상세정보_1.csv'  # 파일 경로를 입력하세요
df = pd.read_csv(csv_file_path)

#df = df.head()

sents = ' '.join(df['overview'].tolist())
sents = df['overview'][0]
print(sents)

# Komoran 초기화
komoran = Komoran()


def komoran_tokenizer(sent):
    words = komoran.pos(sent, join=True)
    words = [w for w in words if ('/NN' in w  )]

    #words = [w for w in words if ('/NN' in w or '/XR' in w )]
    return words

# 텍스트 토큰화
tokenized_words = komoran_tokenizer(sents)

# 토큰화된 결과를 텍스트 파일로 저장
with open('./tale_komoran.txt', 'w', encoding='utf-8') as f:
    f.write(' '.join(tokenized_words))
```

```python
from textrank import KeywordSummarizer

with open("./tale_komoran.txt", "r") as f:
    sents = [sent.strip() for sent in f]

keyword_summarizer = KeywordSummarizer(tokenize=komoran_tokenizer, min_count=2, min_cooccurrence=1)
keywords = keyword_summarizer.summarize(sents, topk=15)

for word, rank in keywords:
     print('{} ({:.3})'.format(word, rank))
```
